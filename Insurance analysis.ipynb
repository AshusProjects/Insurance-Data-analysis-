# Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2: Load Dataset

```python
df = pd.read_csv("C:\----------0000000000----------\insurance.csv")  # Replace with your file path
```

# Step 3: Basic Information

```python
print("Shape of data:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nData Info:")
print(df.info())
print("\nMissing values:")
print(df.isnull().sum())
print("\nStatistical Summary:")
print(df.describe())
```

# Step 4: Univariate Analysis

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], kde=True, bins=20, color='blue')
plt.title("Age Distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df['bmi'], kde=True, color='orange')
plt.title("BMI Distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df['charges'], kde=True, color='green')
plt.title("Charges Distribution")
plt.show()
```

# Step 5: Categorical Features Count

```python
plt.figure(figsize=(6,4))
sns.countplot(x='sex', data=df)
plt.title("Gender Count")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='smoker', data=df)
plt.title("Smoker Count")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='region', data=df)
plt.title("Region Count")
plt.show()
```

# Step 6: Bivariate Analysis

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='smoker', y='charges', data=df)
plt.title("Charges vs Smoker Status")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df)
plt.title("BMI vs Charges (Smoker/Non-Smoker)")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x='age', y='charges', hue='smoker', data=df)
plt.title("Age vs Charges (Smoker/Non-Smoker)")
plt.show()
```

# Step 7: Correlation Heatmap

```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
